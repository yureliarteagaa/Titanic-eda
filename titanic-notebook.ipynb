```python
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('/kaggle/input/titanic-machine-learning-from-disaster/train.csv')

# View the first 5 rows to confirm it's working
titanic.head()

```

```python
titanic['Sex'].value_counts()
```

```python
titanic.groupby('Sex')['Survived'].mean().round(2)
```

```python
titanic['Pclass'].value_counts().sort_index()
```

```python
titanic.groupby(['Pclass'])['Survived'].mean().round(2)
```

```python
titanic['IsChild'] = titanic['Age'] < 18
titanic.groupby('IsChild')['Survived'].mean().round(2)
```
